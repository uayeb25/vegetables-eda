### Ejercicio uno: realice un analisis de valores nulos

En caso de existir valores nulos elimine esos registros

```python
nulldf = pd.DataFrame()
for col_ in vdf.columns:
    r_  = pd.DataFrame({
        'col' : [col_]
        , 'count_nulls' : [sum(pd.isnull(vdf[col_]))]
    })
    nulldf = pd.concat([nulldf, r_])
```

### Ejercicio dos: Por categoria, cuantos distintos items hay y cuantos distintos variety hay?

El resultado debe de tener el siguiente formato de presentacion:

```python
pd.DataFrame({
    "category" : ['cut_flowers','fruit']
    , "item_count" : [9,10]
    , "variety_Count" : [10,21]
})
```

Como nota en un mismo dataframe mostramos todo lo solicitado, usted debe de mostrar todas las categorias

```python
i_ = vdf.loc[ : , ["category","item"] ].drop_duplicates()
v_ = vdf.loc[ : , ["category","variety"] ].drop_duplicates()
```

```python
ic_ =  i_.groupby(["category"],as_index=False).agg({ 'item' : ['count'] })
vc_ =  v_.groupby(["category"],as_index=False).agg({ 'variety' : ['count'] })
```

```python
ic_.columns = ic_.columns.droplevel(1)
vc_.columns = vc_.columns.droplevel(1)
```

```python
ic_.merge( vc_ , on=['category'] , how='inner' )
```

### Ejercicio cuatro: Segun el ejercicio tres, que tanto aumento o disminuyo el precio de cada item-variety 

Ejemplo de como se debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana','Manzana','Manzana']
    , "variety" : ['Gala','Gala','Gala']
    , "year" : [2020,2021,2022]
    , "last_price" :  [2.7,2.5,3.2]
    , "delta" : [0, (2.5-2.7), (3.2-2.5)]
})
```

Segun nuestro ejemplo dummy para 2021 estuvo 0.2 centavos mas barato y en 2022 estuvo 0.7 centavos caro

```python
e4 = e3.copy()
```

```python
e4 = e4.sort_values( ['item','variety','year'] , ascending=True  )
```

```python
e4['prev_price'] = e4.groupby( ['item','variety'] , as_index=False )['price'].shift(1)
```

```python
e4 = e4.loc[ ~pd.isnull(e4['prev_price']) ] 
```

```python
e4['delta'] = e4.apply( lambda r: r['price'] - r['prev_price'] , axis=1 )
```

```python
e4.loc[ : , ['item','variety','year','price','delta'] ]
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
last_values  = vdf.sort_values( [ 'category' , 'item' , 'variety' , 'date' ] , ascending=False )
last_values['r'] = last_values.groupby( ['category' , 'item' , 'variety'] , as_index=False ).cumcount()
last_values = last_values.loc[ last_values['r'] == 0 , ['category' , 'item' , 'variety' , 'price'] ]
last_values.columns = ['category' , 'item' , 'variety' , 'last_price']

first_values  = vdf.sort_values( [ 'category' , 'item' , 'variety' , 'date' ] , ascending=True )
first_values['r'] = first_values.groupby( ['category' , 'item' , 'variety'] , as_index=False ).cumcount()
first_values = first_values.loc[ first_values['r'] == 0 , ['category' , 'item' , 'variety' , 'price'] ]
first_values.columns = ['category' , 'item' , 'variety' , 'first_price']
```

```python
e6 = last_values.merge( first_values , on=['category' , 'item' , 'variety'] , how='inner' )
```

```python
e6['percentage-change-between-first-and-last-price'] =  e6.apply( lambda r : ( (r['last_price']-r['first_price'])/r['first_price'] ) * 100 , axis = 1 ) 
```

```python
e6 = e6.loc[ : , ['category','item','variety', 'percentage-change-between-first-and-last-price']  ]
```

```python
e6
```
